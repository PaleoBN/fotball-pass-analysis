```python
events= events[['team','type','minute','location','pass_end_location','player']]
events = events[events['team']=='Bayer Leverkusen'].reset_index()
```

```python
## stasbomb field x and y: 0 120, 0 80
```

```python
df = df[(df['type']=='Pass')].reset_index()
```

```python
df.head()
```

```python
df.location.dtype
```

```python
df= df[['team','type','location','pass_end_location']]
```

```python
df
```

```python
# Extract start and end coordinates for passes
df['x'] = df['location'].apply(lambda loc: loc[0] if isinstance(loc, list) else None)
df['y'] = df['location'].apply(lambda loc: loc[1] if isinstance(loc, list) else None)
df['endX'] = df['pass_end_location'].apply(lambda loc: loc[0] if isinstance(loc, list) else None)
df['endY'] = df['pass_end_location'].apply(lambda loc: loc[1] if isinstance(loc, list) else None)

# Drop rows with missing coordinates
df = df.dropna(subset=['x', 'y', 'endX', 'endY'])

# Filter out only passes
passes_df = df[['x', 'y', 'endX', 'endY']]
```

```python
passes_df
```

```python
X = np.array(passes_df[['x','y','endX','endY']])
kmeans = KMeans(n_clusters = 5,random_state=100)
kmeans.fit(X)
passes_df['cluster'] = kmeans.predict(X)
```

```python
df.cluster.value_counts()
```

```python
# Plotting only the specified clusters (e.g., clusters 0 and 5) on a soccer pitch
fig, ax = plt.subplots(figsize=(10, 10))
fig.set_facecolor('#38383b')
ax.patch.set_facecolor('#38383b')

pitch = Pitch(pitch_type='statsbomb', pitch_color='#38383b', line_color='white')
pitch.draw(ax=ax)

for idx, row in df.iterrows():
    if row['cluster'] == 3:
        pitch.lines(row['x'], row['y'], row['endX'], row['endY'], color='#74c69d', lw=3, zorder=1, comet=True, ax=ax)
    elif row['cluster'] == 6:
        pitch.lines(row['x'], row['y'], row['endX'], row['endY'], color='#add8e6', lw=3, zorder=1, comet=True, ax=ax)

plt.title('Passes of Bayern Leverkusen vs Werder Bremen on 2024-04-14\n KMeans Clustering of Passes: Clusters 3 and 6  \n Green: Area with the fewes passes, Blue: Area with the most passes' , color='white')
plt.show()
```
